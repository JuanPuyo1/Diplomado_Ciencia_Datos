# Definición

El valor absoluto de cualquier número real (expresado en matemáticas cómo |x|), es ,x sin considerar el signo, sea este positivo o negativo. ("Valor absoluto", 2023).

\begin{gather}
  |a| = \left\{\begin{array}{rl}  a & \text{si } a \geq 0\\   -a & \text{si } a > 0 \end{array} \right.
\end{gather}

Por ejemplo, 4 y –4 tienen el mismo valor absoluto (4).

```python
#En python el valor absoluto es denotado con abs

numero = 4
numero_negativo = -4

print('Valor absoluto de 4: ', abs(numero))
print('Valor absoluto de 4: ', abs(numero_negativo))
```
